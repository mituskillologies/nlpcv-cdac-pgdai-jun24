```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Line Plot

```python
x = np.array([3,11,16,23,35,40,47,56,67,71])
y = np.array([10,16,34,48,61,55,68,72,81,86])
z = np.array([67,66,64,43,32,21,18,14,11,6])
p = np.array([67,55,36,88,64,25,48,62,59,48])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xticks(range(0,100,10))
plt.yticks(range(0,100,10))
plt.xlim(0,100)
plt.ylim(0,100)
# marker: o  .  >   <   ^  v  D  *  +  
plt.plot(x, y, color = 'red', marker = '.', ms=20, mfc='g')
```

```python
plt.plot(y, z, color = 'r', label = '2022')
plt.plot(x, z, color = 'g', label = '2023')
plt.legend()
```

```python
# row, column, figure number
plt.subplot(1,2,1)
plt.plot(x, y)
plt.subplot(1,2,2)
plt.plot(x, z)
```

```python
# row, column, figure number
plt.subplot(2,1,1)
plt.plot(x, y)
plt.subplot(2,1,2)
plt.plot(x, z)
```

```python
# row, column, figure number
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.plot(x, y)
plt.subplot(2,1,2)
plt.plot(x, z)
```

```python
# row, column, figure number
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.plot(x, y, color= 'r')
plt.subplot(1,2,2)
plt.plot(x, z, color = 'g')
plt.subplot(2,2,3)
plt.plot(x, p, color='b')
plt.savefig('myplot.png')
```

```python
df = pd.read_csv('student.csv')
```

```python
plt.plot(df['roll'], df['marks'], color='magenta')
```

#### Scatter plot

```python
plt.scatter(x, y)
```

```python
plt.scatter(df['roll'], df['marks'], color='cyan')
```

#### Bar plot

```python
plt.figure(figsize=(16,9))
plt.title('The marks of students')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(df['name'], df['marks'], color = 'gray')
```

```python
help(plt.bar)
```

#### Horizontal Barplot

```python
plt.figure(figsize=(16,9))
plt.title('The marks of students')
plt.ylabel('Student Name')
plt.xlabel('Marks')
plt.barh(df['name'], df['marks'], color = 'gray')
```

#### Pie chart

```python
df['class'].value_counts()
```

```python
plt.pie(df['class'].value_counts(), labels=['TE','BE'], autopct='%2.2f%%');
```

```python
plt.pie(df['age'].value_counts(), 
        labels= df['age'].value_counts().index, 
        autopct='%2.2f%%', shadow=True, radius=2);
```

#### Histogram

```python
plt.title('Histogram of Marks')
plt.xlabel('Range of Marks')
plt.ylabel('Count')
plt.xticks(range(0,100,10))
plt.grid()
plt.hist(df['marks'], bins=range(0,100,10));
```

#### Box plot

```python
plt.xlabel('Marks')
plt.ylabel('Data Distribution')
plt.grid()
plt.boxplot(df['marks']);
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,3,1)
plt.plot(x, y)
plt.subplot(2,3,2)
plt.scatter(df['roll'], df['marks'], color='cyan')
plt.subplot(2,3,3)
plt.bar(df['name'], df['marks'], color = 'gray')
plt.subplot(2,3,4)
plt.hist(df['marks'], bins=range(0,100,10));
plt.subplot(2,3,5)
plt.pie(df['age'].value_counts(), 
        labels= df['age'].value_counts().index, 
        autopct='%2.2f%%', shadow=True);
plt.subplot(2,3,6)
plt.boxplot(df['marks']);
plt.savefig('allplots.png')
```
