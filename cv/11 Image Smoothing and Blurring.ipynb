##### Averaging Kernel

```python
import cv2
import numpy as np

# Load an image
image = cv2.imread("images/lenna.png")

# Define a 5x5 averaging kernel
kernel = np.ones((5, 5), np.float32) / 25

# Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(image, -1, kernel)

# Show the original and smoothed images
cv2.imshow("Original", image)
cv2.imshow("Smoothed", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image = np.hstack([image, smoothed_image])
```

```python
cv2.imshow("Both images", new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Gaussian Kernel

```python
kernel_size = 7  # Adjust as needed
sigma = 0.35 * ((kernel_size - 1) * 0.5 - 1) + 0.9  # Gaussian standard deviation
kernel = cv2.getGaussianKernel(kernel_size, sigma)
```

```python
sigma
```

```python
kernel
```

```python
# Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(image, -1, kernel)

# Show the original and smoothed images
cv2.imshow("Original", image)
cv2.imshow("Smoothed", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image = np.hstack([image, smoothed_image])
cv2.imshow("Both images", new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Median Filter

```python
gray = cv2.imread('images/lenna.png', 0)
```

```python
gray.shape
```

```python
import matplotlib.pyplot as plt
```

```python
kernel_size = 5
median = cv2.medianBlur(gray, kernel_size)
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(gray, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```

```python
kernel_size = 15
median = cv2.medianBlur(gray, kernel_size)
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(gray, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```

```python
gray = cv2.imread('images/lenna.png')
kernel_size = 15
median = cv2.medianBlur(gray, kernel_size)
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(gray, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title('Median Filtered Image')
plt.show()
```

```python
original_image = cv2.imread('images/lenna.png')

# Define the coordinates of the patch (top-left and bottom-right corners)
x1, y1 = 100, 100
x2, y2 = 300, 300

# Extract the patch from the original image
patch = original_image[x1:x2, y1:y2]

# Apply median blur to the patch
blurred_patch = cv2.medianBlur(patch, ksize=25)

# Replace the original patch with the blurred patch
original_image[x1:x2, y1:y2] = blurred_patch

# Display the resulting image
cv2.imshow('Blurred Image', original_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
