###### Using default rotation

```python
import cv2

# Load the image
image = cv2.imread("images/lenna.png")

# Rotate the image 45 degrees clockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

# Show the rotated image
cv2.imshow("Rotated Image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

###### Rotation using matrix

```python
import cv2

# Load the image
image = cv2.imread("images/lenna.png")

# Define the center point and rotation angle
center = (image.shape[1] // 2, image.shape[0] // 2)
angle = 60

# Create the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Apply the rotation using warpAffine
rotated_image = cv2.warpAffine(image, rotation_matrix, 
                               (image.shape[1], image.shape[0]))

# Show the rotated image
cv2.imshow("Rotated Image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Load the image
image = cv2.imread("images/lenna.png")

# Define the center point and rotation angle
angle = 60

# Create the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D((255,255), angle, 1.0)

# Apply the rotation using warpAffine
rotated_image = cv2.warpAffine(image, rotation_matrix, 
                               (768, 768))

# Show the rotated image
cv2.imshow("Rotated Image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
