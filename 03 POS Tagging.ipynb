```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
text = "Venus is the second planet from the Sun. It is a terrestrial planet and is the closest in mass and size to its orbital neighbour Earth. Venus is notable for having the densest atmosphere of the terrestrial planets, composed mostly of carbon dioxide with a thick, global sulfuric acid cloud cover. At the surface it has a mean temperature of 737 K (464 °C; 867 °F) and a pressure of 92 times that of Earth's at sea level. These extreme conditions compress carbon dioxide into a supercritical state close to Venus's surface."
```

```python
text
```

```python
tags = pos_tag(word_tokenize(text))
```

```python
tags
```

```python
# Extract all the nouns from text and store in list 'nouns'
```

```python
nouns = [n[0] for n in tags if n[1].startswith('N')]
```

```python
nouns
```

```python
# Extract all the adjective-noun pairs from text and store in list 'pairs'
```

```python
pairs = []
for count in range(len(tags)):
    if tags[count][1].startswith('JJ') and tags[count+1][1].startswith('NN'):
        pairs.append([tags[count][0], tags[count+1][0]])
```

```python
pairs
```

##### POS Tagging of Indian Languages

```python
from nltk.corpus import indian
from nltk.tag import TnT
```

```python
for x in indian.fileids():
    print(x)
```

```python
for x in indian.sents():
    print(x)
```

```python
for x in indian.fileids():
    print(x)
    print(len(indian.words(x)))
```

```python
indian.sents('marathi.pos')
```

```python
pos = indian.tagged_sents('marathi.pos')
```

```python
pos
```

```python
# create the object of tagger
tagger = TnT()
```

```python
# train the tagger with tagged sentences
tagger.train(pos)
```

```python
sent = 'सोमवारी हा सामना सुमारे आठ तास उशिरा सुरू झाला.'
```

```python
tagger.tag(word_tokenize(sent))
```
