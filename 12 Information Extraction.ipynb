```python
from nltk.tokenize import word_tokenize
from nltk import FreqDist
```

```python
tokens = word_tokenize(text)
freq = FreqDist(tokens)
freq.plot(20)
```

```python
from nltk.corpus import stopwords
swords = stopwords.words('english')
```

```python
tokens1 = [token for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]
```

```python
freq = FreqDist(tokens2)
freq.plot(20)
```

```python
freq.most_common(10)
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
ps = PorterStemmer()
wnl = WordNetLemmatizer()
```

```python
url = urlopen('https://mr.wikipedia.org/wiki/%E0%A4%B5%E0%A4%BE%E0%A4%98')
data = url.read()
soup = BeautifulSoup(data, 'html.parser')
text = soup.get_text(strip=True)
tokens = word_tokenize(text)

tokens1 = [wnl.lemmatize(token, pos='n') for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]

freq = FreqDist(tokens2)
#freq.plot(20)
freq.most_common(5)
```
