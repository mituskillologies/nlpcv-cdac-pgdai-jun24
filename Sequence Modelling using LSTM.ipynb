```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
int_to_char = dict((i,c) for i, c in enumerate(alphabets))
```

```python
int_to_char;
```

```python
char_to_int = dict((c,i) for i, c in enumerate(alphabets))
```

```python
char_to_int
```

```python
seq_length = 1
x = []
y = []
for i in range(0, len(alphabets)-seq_length, 1):
    seq_in = alphabets[i:i+seq_length]
    seq_out = alphabets[i+seq_length]
    x.append([char_to_int[char] for char in seq_in])
    y.append(char_to_int[seq_out])
    print(seq_in,'->',seq_out)
```

```python
x
```

```python
y
```

```python
import numpy as np
```

```python
X = np.reshape(x, (len(x), seq_length, 1))
```

```python
X.shape
```

```python
X
```

```python
X = X / float(len(alphabets))
```

```python
X
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
```

```python
y_new.shape
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
# create the model
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(y_new.shape[1], activation='softmax'))
```

```python
# compile the model
model.compile(loss='categorical_crossentropy', metrics=['accuracy'],
              optimizer='adam')
```

```python
model.fit(X, y_new, epochs=500, batch_size=1)
```

```python
model.evaluate(X, y_new, batch_size=1)
```

```python
new = 'E'
new = char_to_int[new]     # Extract int value of char
new = np.reshape(new, (1,1,1))  # reshape it
new = new/len(alphabets)   # rescale it
int_to_char[np.argmax(model.predict(new))]  # perform prediction
```
